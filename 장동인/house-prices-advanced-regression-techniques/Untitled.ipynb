```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('default')
```

```python
def sc_chart(feature):
    train.plot(kind = 'scatter', x = feature, y = 'SalePrice', c = 'coral',
              s = 10, figsize = (30,5))
    plt.title('chart')
    plt.show()
```

```python
sc_chart('Neighborhood')
```

```python
import seaborn as sns
```

```python
corr=train.corr()
top_corr=train[corr.nsmallest(79,'SalePrice')['SalePrice'].index].corr()
figure, ax1 = plt.subplots(nrows=1, ncols=1)
figure.set_size_inches(25,15)
sns.heatmap(top_corr, annot=True, ax=ax1)
```
